# Pandas Series

A series is a one-dimensional labeled array capable of holding any data type. The axis labels are collectively called the index.

# Creating series

There are muliple ways to create a pandas series


```python
import pandas as pd
```

```python
labels = ['a','b','c']
my_list = [10,20,30]
dict = {'a':10, 'b':10, 'c':30}
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list, index=labels)
```

```python
pd.Series(dict)
```
